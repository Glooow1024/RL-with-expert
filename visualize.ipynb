```python
import torch
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

file = './results/ExpertDDPG_HalfCheetah-v2_2.npy'
values = np.load(file)
steps = []
value_gd = []
value_pred = []
total_step = 0
for i in range(len(values)):
    step, value = values[i]
    steps += [step]
    value_gd += [value[0]]
    value_pred += [value[1]]
    
plt.plot(steps, value_gd, label='True DDPG')
plt.plot(steps, value_pred, label='Ours DDPG')
plt.legend(loc=4)
```

```python
import torch
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

file = './results/ExpertDDPG_HalfCheetah-v2_3_without_expert.npz'
f = np.load(file)
steps = f['value_step']
value_gd = np.mean(f['value_true'],1)
value_pred = np.mean(f['value_pred'],1)
    
plt.plot(steps, value_gd, label='True DDPG')
plt.plot(steps, value_pred, label='Ours DDPG(NoExpert)')
plt.legend(loc=4)
```
